```python
import pandas as pd
df = pd.read_csv("/Users/saumyagupta/Documents/AI Engineer Roadmap with Projects/Natural Language Processing/Datasets/SMSSpamCollection.txt", sep='\t', names=['label', 'message'])
```

```python
df
```

```python
## Data Cleaning and Preprocessing

import re 
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

stemmer =PorterStemmer()
```

```python
corpus=[]
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['message'][i])
    review = review.lower()
    review = review.split()
    review = [stemmer.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bag_of_words = CountVectorizer(max_features=2500, binary=True)
X = bag_of_words.fit_transform(corpus).toarray()
print(X)
print(X.shape)
```

#### N-Grams

```python
#top 2500 words
bag_of_words.vocabulary_
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bag_of_words = CountVectorizer(max_features=2500, binary=True, ngram_range=(1,1))
X = bag_of_words.fit_transform(corpus).toarray()
print(X)
print(X.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bag_of_words = CountVectorizer(max_features=2500, binary=True, ngram_range=(1,2))
X = bag_of_words.fit_transform(corpus).toarray()
print(X.shape)
X
```

```python
bag_of_words.vocabulary_
```

```python
# Here you can see that the vocabulary have words in the form of 1-grams and 2-grams
# 1-grams are the words themselves and 2-grams are the words that are joined together
# For example, 'hello world' is a 2-gram and 'hello' and 'world' are 1-grams
# So, the vocabulary will have 'hello', 'world', 'hello world' etc.
```
