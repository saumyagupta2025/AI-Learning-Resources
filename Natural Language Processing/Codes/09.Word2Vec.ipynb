```python
import gensim
from gensim.models import Word2Vec, KeyedVectors

import gensim.downloader as api

wv = api.load('word2vec-google-news-300')

```

```python
wv.most_similar('india')
```

```python
wv.similarity('india', 'mango')
```

```python
vec_king = wv['king']
vec_queen = wv['queen']
vec_man = wv['man']
vec_woman = wv['woman']

vec = vec_king - vec_man + vec_woman
print(vec)
```

```python
wv.most_similar(vec)
```
